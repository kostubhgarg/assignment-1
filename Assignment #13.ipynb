```python
#Question 3->
(i) ZeroDivisionError- Raised when the second argument of a division or modulo operation is zero.
Example-> 
```

```python
try:
    num1 = int(input("Enter First Number: "))
    num2 = int(input("Enter Second Number: "))

    result = num1 / num2

    print(result)
except ZeroDivisionError as e:
    print(e)
```

```python
(ii) OverflowError- Raised when the result of an arithmetic operation is too large to be represented. 
```

```python
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
        print(j)
except OverflowError as e:
    print(e)
```

```python
#Question 4->
The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key 
is not found for a sequence or dictionary respectively.
```

```python
(i) Keyerror->
A Python KeyError exception is what is raised when you try to access a key that isn’t in a dictionary (dict).
Example->
```

```python
try:
    d = {"key":1, "k2":2}
    print(d["key2"])
except KeyError as e:
    print(e)
```

```python
(ii) IndexError-> When index is out of bound
Example->
```

```python
try:
    l=[1,2,4]
    print(l[10])
except IndexError as e:
    print(e)
```

```python
#Question 5->
Python’s ImportError (ModuleNotFoundError) indicates that you tried to import a module that Python doesn’t find.
Example->
```

```python
try:
    import koshu
except ImportError as e:
    print(e)
```

```python
try:
    import garg
except ModuleNotFoundError as e:
    print(e)
```
