```python
#Question 3
n Flask, app routing is the process of mapping URLs to view functions, which handle requests from clients and return a 
response. App routes are used to define the URLs that the application will respond to, and to specify which view function 
should handle each URL.

We use app routes in Flask to create a user-friendly interface for our web application. By defining app routes, we can map
specific URLs to specific functions or pages, allowing users to easily navigate our application and access the content they
need.
```

```python
#Question 4
from flask import Flask
from flask import request

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "Company Name: ABC Corporation <br> Location: India <br> Contact Detail: 999-999-9999"

@app.route("/welcome")
def welcomeP():
    return "<h1>Welcome to ABC Corporation</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
#Question 5
The url_for() function is used in Flask to dynamically generate URLs for a given view function or endpoint. 
This function takes the name of the view function or endpoint as its argument and returns a URL that corresponds to 
that function or endpoint.
```

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'This is the home page.'

@app.route('/about')
def about():
    return 'This is the about page.'

@app.route('/profile/<username>')
def profile(username):
    return f'This is {username}\'s profile page.'

with app.test_request_context():
    print(url_for('home'))
    print(url_for('about'))
    print(url_for('profile', username='john'))

```
